```python
def acro(area):
    x=''
    area=area.split()
    if len(area) > 1:
        for i in area:
            x+=i[0]
        return x
    else:
        d=area[0][0:3]
        return d
```

```python
df['acronyms']=df.Area.apply(acro)
```

```python
df
```

```python
df=df.astype({'Value':'float'})
```

```python
df
```

```python
df['Country/Commodity']=df.Area+df.Item
```

```python
df
```

```python
production=df.groupby('Element').get_group('Production')
production
```

```python
yld=df.groupby('Element').get_group('Yield')
yld
```

```python
production_=df.pivot_table(index='Year',columns='Country/Commodity',values='Value')
production_
```

```python
yld_=df.pivot_table(index='Year',columns='Country/Commodity',values='Value')
yld_
```

```python
new=pd.concat([production_,yld_],axis=0,keys=['production_','yld_'])
```
